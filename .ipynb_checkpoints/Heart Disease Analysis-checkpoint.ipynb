# 20. Plot Continuous varibale 

```python
data.columns
```

```python
cate_val = []
cont_val = []

for column in data.columns:
    if data[column].nunique() <= 10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cate_val
```

```python
cont_val
```

```python
data.hist(cont_val, figsize = (15,6))
plt.tight_layout() # for overlapping in graphs 
plt.show()
```
